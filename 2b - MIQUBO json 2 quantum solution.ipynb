```python
import json
import dimod
bqm = dimod.BinaryQuadraticModel.from_serializable(json.load(open("/dbfs/tmp/s.json")))
k = 40       # looking for k sensors to be placed
bqm.update(dimod.generators.combinations(bqm.variables, k, strength=5))   # strength : balance between MI and number of sensor requirement
```

```python
from hybrid.reference.kerberos import KerberosSampler
kerberos_sampler = KerberosSampler() 
best = kerberos_sampler.sample(bqm, 
                                qpu_sampler=qpu_sampler, 
                                qpu_reads=10000, 
                                max_iter=1,
                                qpu_params={'label': 'Notebook - Feature Selection'}
                                ).first.sample
best
```

```python
result = [key for (key, value) in best.items() if value == 1]
result
```
